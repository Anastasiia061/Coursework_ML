# 1. Описание данных

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df=pd.read_csv('../input/telecom_users.csv')
df.head(10)
```

```python
#  удалим колонки Unnamed: 0',customerID, т.к. они не пригодятся для построении модели
df.drop(['Unnamed: 0','customerID'], axis=1, inplace=True)
# посмотрим сводную информацию о таблице
df.info()
```

Для дальнейшей работы приведем данные категориальные признаки к  числовым (бинарным)

```python
# приведем данные виду 0-Нет, 1-Да.
# в колонке 'gender' заменим значения 'Female' на 0, а 'Male' на 1
df['gender'] = df['gender'].apply(lambda x: 1 if x == 'Female' else 0)

# в колонке 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling' приведем данные виду 0-'Yes', 1-'No'.
df['Partner']= df['Partner'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Dependents']= df['Dependents'].apply(lambda x: 1 if x == 'Yes' else 0)
df['PhoneService']= df['PhoneService'].apply(lambda x: 1 if x == 'Yes' else 0)
df['PaperlessBilling']= df['PaperlessBilling'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Churn']= df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# в колонке  'OnlineSecurity', 'OnlineBackup', 'DeviceProtection','TechSupport', 'StreamingTV', 
# 'StreamingMovies' приведем данные к виду 1-'Yes', 0-'No', 0.5-'No internet service'.

def change_name (x):
        if x=='Yes':
            return 1
        elif x=='No':
            return 0
        elif x=='No internet service':
            return 0.5      
           
df['OnlineSecurity']=df.loc[:,'OnlineSecurity'].astype('str').apply(change_name)           
df['OnlineBackup']=df.loc[:,'OnlineBackup'].astype('str').apply(change_name)   
df['DeviceProtection']=df.loc[:,'DeviceProtection'].astype('str').apply(change_name)   
df['TechSupport']=df.loc[:,'TechSupport'].astype('str').apply(change_name)   
df['StreamingTV']=df.loc[:,'StreamingTV'].astype('str').apply(change_name)   
df['StreamingMovies']=df.loc[:,'StreamingMovies'].astype('str').apply(change_name) 


# в колонке 'InternetService' приведем данные к виду 0-'DSL', 1-'Fiber optic', 2-'No'.
def change_internetService (x):
        if x=='DSL':
            return 0
        elif x=='Fiber optic':
            return 1
        elif x=='No':
            return 2 
        
df['InternetService']=df.loc[:,'InternetService'].astype('str').apply(change_internetService) 




# в колонке 'Contract' приведем данные к виду 0-'Month-to-month', 1-'One year', 2-'Two year'.
def change_contract (x):
        if x=='Month-to-month':
            return 0
        elif x=='One year':
            return 1
        elif x=='Two year':
            return 2 
        
df['Contract']=df.loc[:,'Contract'].astype('str').apply(change_contract) 


# в колонке 'MultipleLines' приведем данные к виду 1-'Yes', 0-'No', 2-'No phone service'.
def change_multipleLines (x):
        if x=='Yes':
            return 1
        elif x=='No':
            return 0
        elif x=='No phone service':
            return 0.5  
        
df['MultipleLines']=df.loc[:,'MultipleLines'].astype('str').apply(change_multipleLines) 



# в колонке 'PaymentMethod' приведем данные к виду 0-'Electronic check', 1-'Mailed check',
# 2-Bank transfer (automatic), 3-Credit card (automatic)).
def change_paymentMethod (x):
        if x=='Electronic check':
            return 0
        elif x=='Mailed check':
            return 1
        elif x=='Bank transfer (automatic)':
            return 2
        elif x=='Credit card (automatic)':
            return 3
        
df['PaymentMethod']=df.loc[:,'PaymentMethod'].astype('str').apply(change_paymentMethod) 
# в колонке TotalCharges заменим значения ' ' на медианное значение TotalCharges
df['TotalCharges']=df.loc[:,'TotalCharges'].astype('str').apply(lambda x: 0 if x == ' ' else float(x))
df['TotalCharges']=df.loc[:,'TotalCharges'].apply(lambda x: df['TotalCharges'].median() if x == 0 else x)
```

# 3. Построение моделей для прогнозирования оттока

Для построения модели используем стекинг (ансамблирование алгоритмов). Он  использует разнородные базовые алгоритмы, ошибки одних алгоритмов будут компенсироваться корректной работой других. Так же в данном случае отток в несколько раз меньше,чем общее количество остающихся клиентов, и использование только одного базового алгоритма может дать  в итоге некорректную модель.    
В  качестве мета-алгоритма  используем XGBoost, в качестве базовых алгоритмов -  GradientBoostingClassifier, RandomForestClassifier, kNN, LogisticRegression, DesisionTreeClassifier, SupportVectorClassification и GaussianNB.

```python
# подготовим исходные данные. Размер  выборки возьмем 0.3, random_state=17
import warnings
warnings.filterwarnings('ignore')


from sklearn.neighbors import KNeighborsClassifier 
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
targets = df['Churn']
data=df.drop(['Churn'], axis=1)
train, test, train_true, test_true = train_test_split(data, 
                                                    targets,
                                                    train_size=0.3,
                                                    random_state=17)

x_train, valid, y_train, valid_true = train_test_split(train, 
                                                    train_true,
                                                    train_size=0.7,
                                                    random_state=17)
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score

GBC  = GradientBoostingClassifier(learning_rate=0.13,random_state=17)
RFC = RandomForestClassifier(min_samples_leaf=6, n_estimators=40,random_state=17)
SV = SVC(C= 0.5000000000000001, kernel= 'linear',random_state=17)
LR = LogisticRegression(C =0.7999999999999999, random_state=17)
KNN = KNeighborsClassifier(metric = 'manhattan', n_neighbors=10)
DT = DecisionTreeClassifier(max_depth =7, min_samples_split = 70, random_state=17)

    
nb = GaussianNB()

meta = XGBClassifier(learning_rate = 0.80, min_child_weight = 4, n_estimators= 80, subsample= 0.1, random_state=17)

models=[GBC, RFC, LR, SV, KNN, DT, nb]



meta_mtrx = np.empty((valid.shape[0], len(models)))            
for n, model in enumerate(models):
    model.fit(x_train, y_train)
    meta_mtrx[:, n] = model.predict(valid)
    
    
    
meta.fit(meta_mtrx, valid_true)
meta_mtrx_test = np.empty((test.shape[0], len(models)))

    
for n, model in enumerate(models):
    meta_mtrx_test[:, n] = model.predict(test) 
    
meta_predict = meta.predict(meta_mtrx_test) 
```

# 4. Сравнение качества моделей 

Так как отток в несколько раз меньше,чем общее количество остающихся клиентов, то в данном случае метрика accuracy не будет показательна.   
Для оценки точности моделей найдем метрики, которые используются при бинарной классификации: recall (полнота), precision (точность), f1, AU.


```python
from sklearn.metrics import roc_auc_score, precision_score, f1_score, recall_score

# добавим все полученные значения в датафрейм.
roc_auc=[roc_auc_score(test_true, meta_predict)]
recall =[recall_score(test_true, meta_predict)]
precision = [precision_score(test_true, meta_predict)]
f1 = [f1_score(test_true, meta_predict)]

for i in models:
    roc_auc.append(roc_auc_score(test_true, i.predict(test)))
    recall.append(recall_score(test_true, i.predict(test)))
    precision.append(precision_score(test_true, i.predict(test)))
    f1.append(f1_score(test_true, i.predict(test)))

   
    
Result=pd.DataFrame({'models': ['XGBClassifier','GradientBoostingClassifier','RandomForestClassifier','LogisticRegression','SVC','KNeighborsClassifier','DecisionTreeClassifier','GaussianNB'],
                     'roc_auс_score': roc_auc, 'recall_score': recall, 'precision_score': precision, 'f1_score': f1})

Result=Result.set_index('models')
Result
```

```python
# отобразим сравнение качества моделей на графике
Result.plot(kind='bar')
plt.legend(loc="center right",bbox_to_anchor=(1.4, 0.5))
plt.xlabel('models')
plt.ylabel('score')
plt.show()
```

В нашем случае метрика полноты (recall) важнее,чем точность (важнее  определить какая доля ушедших клиентов была верно определена, но при этом были те, кто ошибочно был определен как "ушедший").      
Исходя из сравнения точности моделей наилучший roc_auс_score, f1_score и recall у  XGBClassifier и GaussianNB - эти модели точнее предскажут уход клиента (без учета признаков  'TotalCharges' и 'Contract')
